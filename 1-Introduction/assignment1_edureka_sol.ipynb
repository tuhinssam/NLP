```python
import string
allAlphabets=list(string.ascii_uppercase)
```

```python
frequency=[]
for each in allAlphabets:
    wordswithEach=filter(lambda x:x.startswith(each),names.words())
    frequency.append(len(wordswithEach))

for every in zip(allAlphabets,frequency):
    print every
```

```python
import matplotlib.pyplot as plt
ticks=range(1,len(allAlphabets)+1)
plt.bar(ticks,frequency)
plt.xticks(ticks,allAlphabets)
plt.show()
```
