ABC Company wants to perform some text analysis andmake visualizationfor one of their dataset.The dataset has been taken from Kaggle. (https://www.kaggle.com/crowdflower/twitter-airline-sentiment/home)This is a dataset having tweets about 6 US Airlines along with their sentiments: positive, negative and neutral.You are provided with this dataset named ‚ÄúTweets.csv‚Äù. This is a snapshot of the dataset:It has tweets in ‚Äòtext‚Äô column and sentiments in ‚Äòairline_sentiment‚Äô column.

1. Retrieve all tags starting from ‚Äò@‚Äô in the entire dataset and save in a file called ‚ÄúReferences.txt‚Äù

2. Extract all noun phrases from their dataset and save them in different lines in a file named ‚ÄúNoun Phrasesfor <airline_sentiment> Review .txt‚Äù (You can choose your own grammar for noun phrase).Here <airline_sentiment> will have 3 different values: positive, negative and neutral so 3 files will be created.

3. Extract all verb phrases from their dataset and save them in different lines in a file named ‚ÄúVerb Phrases for <airline_sentiment> Review .txt‚Äù‚Äù (You can choose your own grammar for noun phrase).Here <airline_sentiment> will have 3 different values: positive , negative and neutral so 3 files will be created.

```python
import pandas as pd
import numpy as np
import os
from nltk.tokenize import word_tokenize
import nltk
```

```python
df = pd.read_csv('Dataset/698_m3_datasets_v1.0/Tweets.csv')
```

```python
df.sample(5)
```

```python
df['text'] = df['text'].astype('str')
```

```python
file1 = open('References.txt','w+',encoding="utf-8")
for item in df['text'].iteritems():
    word_tokens = word_tokenize(item[1])
    cnt = 0
    for word in word_tokens:
        cnt += 1
        if len(word_tokens) > 0:
            if word == "@":
                file1.write(word_tokens[cnt]+ " ")        
file1.close()
```

```python
file1 = open('Noun Phrases for Review-pos.txt','w+',encoding="utf-8")
file2 = open('Noun Phrases for Review-neg.txt','w+',encoding="utf-8")
file3 = open('Noun Phrases for Review-neutral.txt','w+',encoding="utf-8")

for item in df[df["airline_sentiment"]=='positive']['text']:
    word_tokens = word_tokenize(item)
    postags = nltk.pos_tag(word_tokens)
    cnt = 0
    for word,pos in postags:
            if pos.startswith("N"):
                if word == "@":
                    word.replace("@","")
                else:
                    file1.write(word+" ")
file1.close()

for item in df[df["airline_sentiment"]=='negative']['text']:
    word_tokens = word_tokenize(item)
    postags = nltk.pos_tag(word_tokens)
    cnt = 0
    for word,pos in postags:
            if pos.startswith("N"):
                if word == "@":
                    word.replace("@","")
                else:
                    file2.write(word+" ")
file2.close()

for item in df[df["airline_sentiment"]=='neutral']['text']:
    word_tokens = word_tokenize(item)
    postags = nltk.pos_tag(word_tokens)
    cnt = 0
    for word,pos in postags:
            if pos.startswith("N"):
                if word == "@":
                    word.replace("@","")
                else:
                    file3.write(word+" ")
file3.close()
```
